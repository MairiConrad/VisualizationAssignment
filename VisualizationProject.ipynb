# MATLAB- Style

```python
#This visualization shows that the miles per gallon increase over the time period of 1970-1982

plt.bar(years, miles_p_gallon, edgecolor = 'violet', color = 'purple', align = 'center')
plt.xlabel('Model Year', fontsize = 20, color = 'violet', loc='center')
plt.ylabel('Miles Per Gallon', fontsize = 20, color = 'violet')
plt.title('MPG By Year of Car', color = 'violet', fontsize = 25)
plt.yticks(size = 15, color = 'w')
plt.xticks(size = 15, color = 'w')
plt.ylim(bottom = 0, top = 50);
```

```python
#Here you can see that the weight of the cars also decrease over the same time period
plt.bar(years, weight, edgecolor = 'violet', color = 'purple', align = 'center')
plt.xlabel('Model Year', fontsize = 20, color = 'violet', loc='center')
plt.ylabel('Weight of Car', fontsize = 20, color = 'violet')
plt.title('Weight of Car by Year', color = 'violet', fontsize = 25)
plt.yticks(size = 15, color = 'w')
plt.xticks(size = 15, color = 'w');
```

## Make It Object Oriented

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.bar(years, miles_p_gallon, edgecolor = 'violet', color = 'purple', align = 'center')
axes.set_xlabel('Model Year', fontsize = 20, color = 'violet', loc='center')
axes.set_ylabel('Miles Per Gallon', fontsize = 20, color = 'violet')
axes.set_ylim(bottom = 0, top = 50)

plt.yticks(size = 15, color = 'w')
plt.xticks(size = 15, color = 'w')
fig.suptitle('MPG By Year of Car', color = 'violet', fontsize= 25);
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.bar(years, weight, edgecolor = 'violet', color = 'purple', align = 'center')
axes.set_xlabel('Model Year', fontsize = 20, color = 'violet', loc='center')
axes.set_ylabel('Weight of car', fontsize = 20, color = 'violet')

plt.yticks(size = 15, color = 'w')
plt.xticks(size = 15, color = 'w')
fig.suptitle('Weight of Car by Year', color = 'violet', fontsize= 25);
```
